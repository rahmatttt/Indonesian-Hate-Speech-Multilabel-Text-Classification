# Feature Extraction

## Import Packages and Dataset

```python
import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer

import warnings
warnings.filterwarnings("ignore")
```

```python
data_train = pd.read_csv("Data/data_train.csv",sep=";")
data_test = pd.read_csv("Data/data_test.csv",sep=";")
```

```python
data_train.head()
```

```python
data_test.head()
```

## TF-IDF

```python
X_train = data_train['Tweet_Parsed']
X_test = data_test['Tweet_Parsed']
y_train = data_train.drop(['Tweet','Tweet_Parsed'],axis=1)
y_test = data_test.drop(['Tweet','Tweet_Parsed'],axis=1)
```

```python
# Parameter election
ngram_range = (1,1)
min_df = 1
max_df = 1.
max_features = 300
```

```python
tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=ngram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2')
#                         sublinear_tf=True)
```

```python
features_train = tfidf.fit_transform(X_train).toarray()
labels_train = y_train
print(features_train.shape)
```

```python
features_test = tfidf.transform(X_test).toarray()
labels_test = y_test
print(features_test.shape)
```

```python
fitur = tfidf.get_feature_names()
target = data_train.drop(['Tweet','Tweet_Parsed'],axis = 1).columns
```

```python
data_train_tfidf = pd.DataFrame(data=features_train,columns=fitur)
data_train_tfidf = data_train_tfidf.join(pd.DataFrame(data=y_train,columns=target))
data_train_tfidf.head()
```

```python
data_test_tfidf = pd.DataFrame(data=features_test,columns=fitur)
data_test_tfidf = data_test_tfidf.join(pd.DataFrame(data=y_test,columns=target))
data_test_tfidf.head()
```

## Save File

```python
data_train_tfidf.to_csv('Data/data_train_tfidf.csv', index = False)
data_test_tfidf.to_csv('Data/data_test_tfidf.csv', index = False)
```
