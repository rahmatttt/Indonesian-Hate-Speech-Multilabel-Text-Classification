## Top 20 Words

```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

def get_top_n_bigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2, 2)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
plt.figure(figsize=(20,12))
plt.suptitle("top 20 words")

plt.subplot(2,2,1)
plt.subplots_adjust(hspace=0.4)
common_words_ = get_top_n_words(data_anl_train_['Tweet_Parsed'], 20)
# for word, freq in common_words:
#     print(word, freq)
topWord_ = pd.DataFrame(common_words_, columns = ['Tweet_Parsed' , 'count'])
topWord_.groupby('Tweet_Parsed').sum()['count'].sort_values(ascending=False).plot(
    kind='bar', y='Count', color='red')
plt.title("Data Train Without Translation")
plt.xlabel("")

plt.subplot(2,2,2)
plt.subplots_adjust(hspace=0.4)
common_words_ = get_top_n_words(data_anl_test_['Tweet_Parsed'], 20)
# for word, freq in common_words:
#     print(word, freq)
topWord_ = pd.DataFrame(common_words_, columns = ['Tweet_Parsed' , 'count'])
topWord_.groupby('Tweet_Parsed').sum()['count'].sort_values(ascending=False).plot(
    kind='bar', y='Count', color='yellow')
plt.title("Data Test Without Translation")
plt.xlabel("")

plt.subplot(2,2,3)
plt.subplots_adjust(hspace=0.4)
common_words_ = get_top_n_words(data_anl_train_trans_['Tweet_Parsed'], 20)
# for word, freq in common_words:
#     print(word, freq)
topWord_ = pd.DataFrame(common_words_, columns = ['Tweet_Parsed' , 'count'])
topWord_.groupby('Tweet_Parsed').sum()['count'].sort_values(ascending=False).plot(
    kind='bar', y='Count', color='green')
plt.title("Data Train With Translation")
plt.xlabel("")

plt.subplot(2,2,4)
plt.subplots_adjust(hspace=0.4)
common_words_ = get_top_n_words(data_anl_test_trans_['Tweet_Parsed'], 20)
# for word, freq in common_words:
#     print(word, freq)
topWord_ = pd.DataFrame(common_words_, columns = ['Tweet_Parsed' , 'count'])
topWord_.groupby('Tweet_Parsed').sum()['count'].sort_values(ascending=False).plot(
    kind='bar', y='Count', color='blue')
plt.title("Data Test With Translation")
plt.xlabel("")
```

## Label Count

```python
def countLabel(data):
    label_size0 = []
    label_size1 = []

    for i in data.columns[:-1]:
        label_size0.append(len(data[data[i]==0]))
        label_size1.append(len(data[data[i]==1]))
        
    dat_anls = pd.DataFrame({
        'label' : data.columns[:-1],
        'negative' : label_size0,
        'positive' : label_size1
    })

    dat_anls = dat_anls.set_index(['label'], drop=True)
    return dat_anls
```

```python
data_train_anls = countLabel(data_anl_train)
data_test_anls = countLabel(data_anl_test)
data_train_trans_anls = countLabel(data_anl_train_trans)
data_test_trans_anls = countLabel(data_anl_test_trans)
```
