# Feature Extraction


## Import Packages and Data

```python
import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split


import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("Data/Preprocessed_Data.csv")
data.head()
```

```python
pd.isnull(data).sum()
```

## Train Test Split

```python
X = data['Tweet_Parsed']
y = data.drop(['Tweet','Tweet_Parsed'],axis = 1).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=8)
```

## TF-IDF

```python
# Parameter election
ngram_range = (1,1)
min_df = 1
max_df = 1.
max_features = 300
```

```python
tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=ngram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2',
                        sublinear_tf=True)
```

```python
features_train = tfidf.fit_transform(X_train).toarray()
labels_train = y_train
print(features_train.shape)
```

```python
features_test = tfidf.transform(X_test).toarray()
labels_test = y_test
print(features_test.shape)
```

```python
fitur = tfidf.get_feature_names()
target = data.drop(['Tweet','Tweet_Parsed'],axis = 1).columns
```

```python
data_train = pd.DataFrame(data=features_train,columns=fitur)
data_train = data_train.join(pd.DataFrame(data=y_train,columns=target))
data_train.head()
```

```python
data_test = pd.DataFrame(data=features_test,columns=fitur)
data_test = data_test.join(pd.DataFrame(data=y_test,columns=target))
data_test.head()
```

## Save File

```python
data_train.to_csv('Data/Data_train.csv', index = False)
data_test.to_csv('Data/Data_test.csv', index = False)
```
